```python
import requests
import pandas as pd
import random
import re
```

```python
# Парсер вакансий

number_of_pages = 100

specialization = ['Автомобильный бизнес', 'Административный персонал', 'Безопасность', 'Менеджмент', 'Добыча сырья',
                  'Домашний персонал', 'Обслуживающий персонал', 'Закупки', 'Информационные технологии', 'Искусство', 'Развлечения', 'Медиа',
                  'Маркетинг', 'Реклама', 'PR', 'Медицина', 'Фармацевтика', 'Наука', 'Образование', 'Продажи', 'Обслуживание клиентов',
                  'Производство', 'Сервисное обслуживание', 'Рабочий персонал', 'Розничная торговля', 'Сельское хозяйство',
                  'Спортивные клубы', 'Фитнес', 'Салоны красоты', 'Стратегия', 'Инвестиции', 'Консалтинг', 'Страхование',
                  'Строительство', 'Недвижимость', 'Транспорт', 'Логистика', 'Перевозки', 'Туризм', 'Гостиницы', 'Рестораны',
                  'Управление персоналом', 'Тренинги', 'Финансы', 'Бухгалтерия', 'Юристы', 'Дизайн']

for job_title in specialization:
  job_title = [job_title]
  for job in job_title:
      data=[]
      for i in range(number_of_pages):
          url = 'https://api.hh.ru/vacancies'
          par = {'text': job, 'area':'113','per_page':'10', 'page':i}
          r = requests.get(url, params=par)
          e=r.json()
          data.append(e)
          vacancy_details = data[0]['items'][0].keys()
          df = pd.DataFrame(columns= list(vacancy_details))
          ind = 0
          for i in range(len(data)):
              for j in range(len(data[i]['items'])):
                  df.loc[ind] = data[i]['items'][j]
                  ind+=1
      csv_name = job+".csv"
      df.to_csv(csv_name)
```

```python
# Функция для фильтрации по специализации и профессии
def filter_specialization(special):
    return pd.read_csv(special + '.csv')
```

```python
# Функция для фильтрации по опыту
def filter_experience(row, min_years):
    match = re.search(r'(\d+)', row)
    if match:
        years = int(match.group(1))
        return years <= min_years or row.find('Нет опыта') != -1
    return False

#min_experience_years = 3
#filtered_vacancies = df[df['experience'].apply(lambda x: filter_experience(x, min_experience_years))]
#filtered_vacancies.head(10)
```

```python
# Функция для фильтрации по зарплате
def filter_salary(row, min_salary):
    match = re.search(r'(\d+)', row)
    if match:
        salary = int(match.group(1))
        return salary <= min_salary
    return False
```

```python
# Функция для фильтрации по типу занятности
def filter_employment(row, employment):
    return row.find(employment) != -1
```

```python
# Подбор вакансий под резюме работника
vacancies_df = filter_specialization(user['сфера работы'])

filtered_vacancies = vacancies_df[vacancies_df['experience'].apply(lambda x: filter_experience(x, user['опыт работы']))]
filtered_vacancies = vacancies_df[vacancies_df['salary'].apply(lambda x: filter_salary(str(x), user['зарплата']))]
filtered_vacancies = vacancies_df[vacancies_df['employment'].apply(lambda x: filter_employment(x, user['тип занятности']))]

filtered_vacancies.head(10)
```
