```python
import requests
import pandas as pd
import numpy as np
import random
import re
```

```python
# Парсер вакансий

number_of_pages = 100

specialization = ['Автомобильный бизнес', 'Административный персонал', 'Безопасность', 'Менеджмент', 'Добыча сырья',
                  'Домашний персонал', 'Обслуживающий персонал', 'Закупки', 'Информационные технологии', 'Искусство', 'Развлечения', 'Медиа',
                  'Маркетинг', 'Реклама', 'PR', 'Медицина', 'Фармацевтика', 'Наука', 'Образование', 'Продажи', 'Обслуживание клиентов',
                  'Производство', 'Сервисное обслуживание', 'Рабочий персонал', 'Розничная торговля', 'Сельское хозяйство',
                  'Спортивные клубы', 'Фитнес', 'Салоны красоты', 'Стратегия', 'Инвестиции', 'Консалтинг', 'Страхование',
                  'Строительство', 'Недвижимость', 'Транспорт', 'Логистика', 'Перевозки', 'Туризм', 'Гостиницы', 'Рестораны',
                  'Управление персоналом', 'Тренинги', 'Финансы', 'Бухгалтерия', 'Юристы', 'Дизайн']
'''
Бухгалтерия и финансы
Разработка программного обеспечения
Медиа, журналистика и искусство
Компьютерная графика и дизайн
Маркетинг и связи с общественностью
Аналитика и хранение данных
Администрирование и инфраструктура
Управление персоналом
Психология и коучинг
Продажи и сопровождение клиентов
Строительство и инженерия
Юриспруденция
Косметология и здоровье
Производство
Управление и бизнес
Разработка компьютерных игр
Перевозки, транспортная и складская логистика
Наука и образование
Розничная торговля
'''

for job_title in specialization:
  job_title = [job_title]
  for job in job_title:
      data=[]
      for i in range(number_of_pages):
          url = 'https://api.hh.ru/vacancies'
          par = {'text': job, 'area':'113','per_page':'10', 'page':i}
          r = requests.get(url, params=par)
          e=r.json()
          data.append(e)
          vacancy_details = ['id', 'name', 'area', 'salary_from', 'salary_to',
                             'addres', 'employer', 'requirement',
                             'responsibility', 'contacts', 'schedule',
                             'working_days', 'work_format', 'working_hours',
                             'night_shifts', 'professional_roles', 'experience',
                             'employment', 'internship']
          p = [['id'], ['name'], ['area', 'name'], ['salary', 'from'], ['salary', 'to'],
               ['address', 'raw'], ['employer', 'name'], ['snippet', 'requirement'],
               ['snippet', 'responsibility'], ['contacts'], ['schedule', 'name'],
               ['working_days', 'name'], ['work_format', 'name'], ['working_hours', 'name'],
               ['night_shifts'], ['professional_roles', 'name'], ['experience', 'name'],
               ['employment', 'name'], ['internship']]
          df = pd.DataFrame(columns= list(vacancy_details))
          ind = 0
          for i in range(len(data)):
              for j in range(len(data[i]['items'])):
                  d = {}
                  for k in range(len(vacancy_details)):
                      t = p[k]
                      c = vacancy_details[k]
                      if data[i]['items'][j][t[0]] is None:
                          d[c] = np.NaN
                      elif type(data[i]['items'][j][t[0]]) == type([1, 2, 3]):
                          if len(data[i]['items'][j][t[0]]) == 0:
                              d[c] = np.NaN
                          else:
                              arr = [i[t[1]] for i in data[i]['items'][j][t[0]]]
                              d[c] = ', '.join(arr)
                      elif len(t) == 1:
                          d[c] = data[i]['items'][j][t[0]]
                      else:
                          if data[i]['items'][j][t[0]][t[1]] is None:
                              d[c] = np.NaN
                          else:
                              d[c] = data[i]['items'][j][t[0]][t[1]]
                  d['area_work'] = job
                  df.loc[ind] = d
                  ind+=1
      csv_name = job+".csv"
      df.to_csv(csv_name)
```

```python
# Создаем пустой список для хранения DataFrame
dfs = []

# Читаем каждый файл и добавляем его в список
for special in specialization:
    df = pd.read_csv(special + '.csv')
    df['branch'] = special
    dfs.append(df)

# Объединяем все DataFrame в один
combined_df = pd.concat(dfs, ignore_index=True)

# Сохраняем объединенный DataFrame в новый CSV файл
combined_df.to_csv('vacancies.csv', index=False)
```

```python
# Функция для фильтрации по опыту
def filter_experience(row, min_years):
    match = re.search(r'(\d+)', row)
    if match:
        years = int(match.group(1))
        return years <= min_years or row.find('Нет опыта') != -1
    return False

#min_experience_years = 3
#filtered_vacancies = df[df['experience'].apply(lambda x: filter_experience(x, min_experience_years))]
#filtered_vacancies.head(10)
```

```python
# Функция для фильтрации по зарплате
def filter_salary(row, min_salary):
    match = re.search(r'(\d+)', row)
    if match:
        salary = int(match.group(1))
        return salary <= min_salary
    return False
```

```python
# Функция для фильтрации по типу занятности
def filter_employment(row, employment):
    return row.find(employment) != -1
```
